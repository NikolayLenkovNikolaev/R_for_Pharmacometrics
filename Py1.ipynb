```
a <- data.frame(
    "time" = c(0:10),
    "conc" = c(0, 20, 100, 96, 94, 92, 90, 60, 30, 5, 1),
    "dose" = rep(5,11)
)

a$conc <- a$conc * 10

# we have multiplied the "conc" column by 10 and put the
# new values back into the 'a' data frame

a$conc10 <- a$conc * 10
```

````python
time = list(range(0,11))
conc = [0, 20, 100, 96, 94, 92, 90, 60, 30, 5, 1]
dose = [5]*11

dict = {"time": time, "conc": conc, "dose": dose}
df = pd.DataFrame(dict)
df
````

````python
df["conc10"] = df["conc"]*10
df
````
